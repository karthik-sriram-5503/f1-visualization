# **formula one visualization**

```python
import pandas as pd
```

```python
results = pd.read_csv(r'/content/drive/MyDrive/nit project/results.csv')
races = pd.read_csv(r'/content/drive/MyDrive/nit project/races.csv')
quali = pd.read_csv(r'/content/drive/MyDrive/nit project/qualifying.csv')
drivers = pd.read_csv(r'/content/drive/MyDrive/nit project/drivers.csv')
constructors = pd.read_csv(r'/content/drive/MyDrive/nit project/constructors.csv')
circuit = pd.read_csv(r'/content/drive/MyDrive/nit project/circuits.csv')
```

```python
df1 = pd.merge(races,results,how='inner',on=['raceId'])
df2 = pd.merge(df1,quali,how='inner',on=['raceId','driverId','constructorId'])
df3 = pd.merge(df2,drivers,how='inner',on=['driverId'])
df4 = pd.merge(df3,constructors,how='inner',on=['constructorId'])
df5 = pd.merge(df4,circuit,how='inner',on=['circuitId'])
```

```python
df5
```

```python
df5.drop('url_x',axis=1)
```

```python
df5.rename(columns={'name_y':'constructor'},inplace=True)
df5.rename(columns={'nationality_x':'driver_nationality'},inplace=True)
df5.rename(columns={'nationality_y':'constructor_nationality'},inplace=True)
df5.info()
```

```python
data = df5.drop(['url_x','url_y'],1)
```

```python
data.info()
```

```python
column_names_list = data.columns.tolist()
column_names_list
```

```python
data=data[data['year']>=2010]
data.head()
```

```python
#renaming colums
data['driver'] = data['forename']+' '+data['surname']
data['date'] = pd.to_datetime(data['date'])
data['dob'] = pd.to_datetime(data['dob'])
#creating age parameters
data['age_at_gp_in_days'] = abs(data['dob']-data['date'])
data['age_at_gp_in_days'] = data['age_at_gp_in_days'].apply(lambda x: str(x).split(' ')[0])
#renaming constructors names
data['constructor'] = data['constructor'].apply(lambda x: 'Racing Point' if x=='Force India' else x)
data['constructor'] = data['constructor'].apply(lambda x: 'Alfa Romeo' if x=='Sauber' else x)
data['constructor'] = data['constructor'].apply(lambda x: 'Renault' if x=='Lotus F1' else x)
data['constructor'] = data['constructor'].apply(lambda x: 'AlphaTauri' if x=='Toro Rosso' else x)
data.head()
```

```python
active_constructors = ['Renault', 'Williams', 'McLaren', 'Ferrari', 'Mercedes',
                       'AlphaTauri', 'Racing Point', 'Alfa Romeo', 'Red Bull',
                       'Haas F1 Team']
active_drivers = ['Daniel Ricciardo', 'Kevin Magnussen', 'Carlos Sainz',
                  'Valtteri Bottas', 'Lance Stroll', 'George Russell',
                  'Lando Norris', 'Sebastian Vettel', 'Kimi Räikkönen',
                  'Charles Leclerc', 'Lewis Hamilton', 'Daniil Kvyat',
                  'Max Verstappen', 'Pierre Gasly', 'Alexander Albon',
                  'Sergio Pérez', 'Esteban Ocon', 'Antonio Giovinazzi',
                  'Romain Grosjean','Nicholas Latifi']
data['active_driver'] = data['driver'].apply(lambda x: int(x in active_drivers))
data['active_constructor'] = data['constructor'].apply(lambda x: int(x in active_constructors))
```

```python
data= data[(data['active_driver']==1)&(data['active_constructor']==1)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
grouped_data = data.groupby(['constructor', 'year'])['position_y'].mean().reset_index().round()
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Create a line chart for each team with customized markers, colors, and line styles
for team in grouped_data['constructor'].unique():
    team_data = grouped_data[grouped_data['constructor'] == team]
    plt.plot(team_data['year'], team_data['position_y'], marker='o', label=team, linewidth=2, markersize=8)

plt.xlabel('Year', fontsize=14, labelpad=10)
plt.ylabel('Average Finishing Position of drivers', fontsize=14, labelpad=10)
plt.title('Team Average Finishing Positions Over the Years', fontsize=16, pad=20)
plt.xticks(grouped_data['year'].unique(), fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.ylim(0, 20)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
sns.despine(left=True)

# Save the plot as an image (optional)
plt.savefig('team_finishing_positions.png', dpi=300)

plt.show()
```

```python
data['driver_dnf'] = data['statusId'].apply(lambda x: 1 if x in [3,4,20,29,31,41,68,73,81,97,82,104,107,130,137] else 0)
data['constructor_dnf'] = data['statusId'].apply(lambda x: 1 if x not in [3,4,20,29,31,41,68,73,81,97,82,104,107,130,137,1] else 0)
import matplotlib.pyplot as plt
dnf_by_driver = data.groupby('driver').sum()['driver_dnf']
driver_race_entered = data.groupby('driver').count()['driver_dnf']
driver_dnf_ratio = (dnf_by_driver*100/driver_race_entered).sort_values(ascending=False)

# Set a color palette for the bars
colors = plt.cm.viridis(driver_dnf_ratio / float(max(driver_dnf_ratio)))

# Create the bar chart with custom styles
plt.figure(figsize=(15, 8))
bars = plt.bar(driver_dnf_ratio.index, height=driver_dnf_ratio, align='center', width=0.5, color=colors)
plt.xticks(rotation=90)
plt.xlabel('Drivers', fontsize=14)
plt.ylabel('Driver DNF Ratio (%)', fontsize=14)
plt.title('DNFs Ratio Due to Driver Error', fontsize=16, pad=20)
plt.ylim(0, max(driver_dnf_ratio) + 10)  # Set y-axis limit with a buffer
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().set_axisbelow(True)  # Put gridlines behind the bars for better visibility
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add data labels above the bars
for bar, label in zip(bars, driver_dnf_ratio):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.5, f'{label:.2f}%', ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

```python
dnf_by_constructor = data.groupby('constructor').sum()['constructor_dnf']
constructor_race_entered = data.groupby('constructor').count()['constructor_dnf']
constructor_dnf_ratio = (dnf_by_constructor*100/constructor_race_entered).sort_values(ascending=False)
```

```python
import numpy as np

# Set up the color palette from black to red
colors = plt.cm.spring(np.linspace(0, 1, len(constructor_dnf_ratio)))

# Create the bar chart with custom styles
plt.figure(figsize=(15, 8))
bars = plt.bar(constructor_dnf_ratio.index, height=constructor_dnf_ratio, align='center', width=0.8, color=colors)
plt.xticks(rotation=90)
plt.xlabel('Constructors', fontsize=14)
plt.ylabel('Constructor DNF Ratio (%)', fontsize=14)
plt.title('DNFs Ratio Due to Constructor Error', fontsize=16, pad=20)
plt.ylim(0, max(constructor_dnf_ratio) + 10)  # Set y-axis limit with a buffer
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().set_axisbelow(True)  # Put gridlines behind the bars for better visibility
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add data labels above the bars
for bar, label in zip(bars, constructor_dnf_ratio):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.5, f'{label:.2f}%', ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

```python
import folium
# Create a base map centered around a specific location (e.g., New York City)
map_center = [40.7128, -74.0060]  # Latitude and longitude of the map center
map_zoom = 4  # Initial zoom level

# Create a Folium map
map_circuits = folium.Map(location=map_center, zoom_start=map_zoom)

# Add circuit names as markers on the map
for i in range(len(circuit['circuitId'])):
    circuit_name = circuit['name'][i]
    latitude = circuit['lat'][i]
    longitude = circuit['lng'][i]

    folium.Marker([latitude, longitude], popup=circuit_name).add_to(map_circuits)

# Save the map as an HTML file
map_circuits.save('circuits_map.html')
```

```python
df['Count'] = df.groupby('name')['name'].transform('count')
```

```python
min_count = df['Count'].min()
max_count = df['Count'].max()
df['Scaled_Count'] = (df['Count'] - min_count) / (max_count - min_count)
```

```python
import folium
from folium.plugins import HeatMap
map_center = [df['lat'].mean(), df['lng'].mean()]  # Center the map based on the average latitude and longitude
map_heatmap = folium.Map(location=map_center, zoom_start=3)

# Create a list of points for the heatmap with scaled values
heatmap_data = [[row['lat'], row['lng'], row['Scaled_Count']] for index, row in df.iterrows()]

# Add the heatmap layer to the map
HeatMap(heatmap_data, radius=15).add_to(map_heatmap)

# Save the map as an HTML file
map_heatmap.save('scaled_heatmap_map.html')





```
